```
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```
func = lambda th: np.sin( 1/2 * th[0] ** 2 - 1/4 * th[1] ** 2 + 3) * np.cos( 2 * th[0] + 1 - np.e ** th[1] )
```

```
res = 100 #resolution

_x = np.linspace(-3, 3, res)
_y = np.linspace(-4, 2.5, res)
_z = np.zeros((res, res))

func([2,2])
```

```
for ix, x in enumerate(_x):
  for iy, y in enumerate(_y):
    _z[iy, ix] = func([x, y])
```

```
plt.contourf(_x, _y, _z, 100)
plt.colorbar()

Theta = np.random.rand(2) * 4 - 2
_T = np.copy(Theta)
h = 0.001
lr = 0.001 # Learning rate

grad_v = np.zeros(2) # Gradient Vector

init_cords = Theta

for _ in range(10000):
  for it, th in enumerate(Theta):

    delta = []

    _T = np.copy(Theta)

    _T[it] = _T[it] + h

    d = (func(_T) - func(Theta)) / h

    grad_v[it] = d
  
  Theta = Theta - grad_v * lr
  if (_ % 100 == 0):
    plt.plot(Theta[0], Theta[1], "o", c="black")

plt.plot(init_cords[0], init_cords[1], "o", c="red" )
plt.plot(Theta[0], Theta[1], "o", c="green")
plt.show()
```
